```python
import csv
import os
import random  
from numpy import random as npRandom
import numpy as np
import matplotlib.pyplot as plt  
def generateCSV(path,numbers,noise=False):
    data=[]
    for i in range(numbers):
        height=round(random.gauss(174, 5),1)
        weight=round(random.gauss(64, 5),1)
        highestEducation=np.argmax(npRandom.multinomial(n=1, pvals=[0.05, 0.3, 0.45, 0.15, 0.05])+1)
        wages=np.argmax(npRandom.multinomial(n=1, pvals=[0.15, 0.10, 0.50, 0.10, 0.10,0.05])+1)
        drinkingOrSmoke=np.argmax(npRandom.multinomial(n=1, pvals=[0.21, 0.49, 0.09, 0.21])+1)
        Appearance=round(random.gauss(5, 1))
        dressStyle=random.randint(1,3)
        lovers=np.argmax(npRandom.multinomial(n=1, pvals=[0.30, 0.50, 0.15, 0.05])+1)
        sports=np.argmax(npRandom.multinomial(n=1, pvals=[0.30, 0.40, 0.20, 0.05])+1)
        tmp=np.argmax(npRandom.multinomial(n=1, pvals=[0.35, 0.65])+1)
        tatoo=random.randint(1,10) if tmp==0 else 0
        houses=np.argmax(npRandom.multinomial(n=1, pvals=[0.40, 0.40, 0.15, 0.05])+1)
        hookUp=np.argmax(npRandom.multinomial(n=1, pvals=[0.70, 0.25, 0.03, 0.02])+1)
        if noise:
            if highestEducation<=2:
                drinkingOrSmoke=np.argmax(npRandom.multinomial(n=1, pvals=[0.075, 0.425, 0.075, 0.425])+1)
            elif highestEducation>=4 and wages<=5:
                if wages<=5:
                    wages+=1
                tmp=random.randint(0,1)
                if tmp==1:
                    if houses<=3:
                        houses+=1
                if dressStyle==3:
                    if Appearance<=9:
                        Appearance+=1
                if Appearance>=7:
                    tmp=random.randint(1,10)
                    if tmp<=3:
                        if hookUp<=3:
                            hookUp+=1     
        if houses>=3 or wages>=5 or Appearance>=7 or (height>=180 and weight<=75 and sports>=3 and drinkingOrSmoke<=2):
            judgement=2
        elif houses>=2 or wages>=4 or Appearance>=6 or (height>=170 and weight<=70 and sports>=2 and drinkingOrSmoke<=2 and hookUp==1):
            judgement=1
        else:
            judgement=0
        data.append([height,weight,highestEducation,wages,drinkingOrSmoke,Appearance,dressStyle,lovers,sports,tatoo,houses,hookUp,judgement])
        with open(path, 'w') as f:
            writer = csv.writer(f)
            writer.writerow(["height", "weight","highestEducation","wages","drinkingOrSmoke","Appearance","dressStyle",
                        "lovers", "sports", "tatoo","houses","hookUp","judgement"])
            writer.writerows(data)
        
```

```python
#generate data
generateCSV("inputs/NONoiseData.csv",10000)
generateCSV("inputs/NoiseData.csv",10000,True)
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
#split data to training and test set
dataNoNoise=shuffle(dataNoNoise,random_state=0)
dataNoise=shuffle(dataNoise,random_state=0)
trainNoNoise,testNoNoise=train_test_split(dataNoNoise, test_size=0.3)
trainNoise,testNoise=train_test_split(dataNoise, test_size=0.3)
trainDataNoNoise,trainLabelNoNoise=trainNoNoise.drop('judgement',axis=1),trainNoNoise['judgement']
testDataNoNoise,testLabelNoNoise=testNoNoise.drop('judgement',axis=1),testNoNoise['judgement']
trainDataNoise,trainLabelNoise=trainNoise.drop('judgement',axis=1),trainNoise['judgement']
testDataNoise,testLabelNoise=testNoise.drop('judgement',axis=1),testNoise['judgement']
```

```python
from sklearn import tree
#DecisionTree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(trainDataNoNoise, trainLabelNoNoise)
acc=clf.score(testDataNoNoise,testLabelNoNoise)
print(acc)
tree.plot_tree(clf)
```

```python
#DecisionTree
clf1 = tree.DecisionTreeClassifier()
clf1 = clf1.fit(trainDataNoise, trainLabelNoise)
acc1=clf1.score(testDataNoise,testLabelNoise)
print(acc1)
tree.plot_tree(clf1)
```

```python
from sklearn.svm import SVC
#SVM
clf = SVC(gamma='auto')
clf = clf.fit(trainDataNoNoise, trainLabelNoNoise)
acc=clf.score(testDataNoNoise,testLabelNoNoise)
print("NoNoise:",acc)
clf1 = SVC(gamma='auto')
clf1 = clf1.fit(trainDataNoise, trainLabelNoise)
acc1=clf1.score(testDataNoise,testLabelNoise)
print("Noise:",acc1)
```
